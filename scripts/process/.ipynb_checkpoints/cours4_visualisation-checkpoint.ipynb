```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../../data/clean/reviews_with_ratings.csv')

# Calculer la longueur de chaque commentaire
df['length'] = df['commentaire'].apply(len)

# Tracer l'histogramme de la longueur du texte
plt.figure(figsize=(10, 6))
plt.hist(df['length'], bins=50, edgecolor='black')
plt.title('Distribution of Review Lengths')
plt.xlabel('Review Length')
plt.ylabel('Number of Reviews')

plt.show()
```

```python
from collections import Counter

# Extraire les contenus des critiques
reviews_text = df['commentaire']

# Fusionner les textes des critiques
concat_text = ' '.join(reviews_text)

# Diviser le texte en une liste de mots en utilisant les espaces comme sparateurs
words = concat_text.split()

# Calculer la frquence des mots
word_freq = Counter(words)

# Extraire les 150 mots les plus frquents
top_n = 150
top_words = word_freq.most_common(top_n)

# Dessiner la distribution de Zipf
plt.figure(figsize=(10, 6))
plt.plot(range(1, top_n+1), [count for word, count in top_words])
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.title('Mots Frquents (Zipf)')
plt.xticks(range(0, top_n+1, 50))
plt.gca()
plt.show()
```
